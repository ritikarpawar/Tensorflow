```python
import tensorflow as tf
import numpy as np


# Load Fashion-MNIST
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0


# Split data
X_pre = X_train[y_train > 1]      # 8 classes
y_pre = y_train[y_train > 1] - 2

X_bin = X_train[y_train < 2]      # 2 classes
y_bin = y_train[y_train < 2]

X_test_bin = X_test[y_test < 2]
y_test_bin = y_test[y_test < 2]


# Pretrained model
pretrained = tf.keras.Sequential([

    tf.keras.layers.Flatten(
        input_shape=(28,28)
    ),

    tf.keras.layers.Dense(
        100,
        activation='relu'
    ),

    tf.keras.layers.Dense(
        8,
        activation='softmax'
    )
])

pretrained.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

pretrained.fit(
    X_pre,
    y_pre,
    epochs=3,
    verbose=0
)


# Model from scratch
scratch = tf.keras.Sequential([

    tf.keras.layers.Flatten(
        input_shape=(28,28)
    ),

    tf.keras.layers.Dense(
        100,
        activation='relu'
    ),

    tf.keras.layers.Dense(
        1,
        activation='sigmoid'
    )
])

scratch.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

scratch.fit(
    X_bin,
    y_bin,
    epochs=3,
    verbose=0
)

_, scratch_acc = scratch.evaluate(
    X_test_bin,
    y_test_bin,
    verbose=0
)

print("\nScratch Accuracy:", scratch_acc)


# Transfer Learning
transfer = tf.keras.Sequential([

    pretrained.layers[0],

    pretrained.layers[1],

    tf.keras.layers.Dense(
        1,
        activation='sigmoid'
    )
])


# Freeze pretrained layer
transfer.layers[1].trainable = False


transfer.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

transfer.fit(
    X_bin[:5000],     # less data
    y_bin[:5000],
    epochs=3,
    verbose=0
)

_, transfer_acc = transfer.evaluate(
    X_test_bin,
    y_test_bin,
    verbose=0
)

print("Transfer Learning Accuracy:", transfer_acc)
```
