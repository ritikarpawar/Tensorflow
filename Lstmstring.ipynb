```python
import numpy as np
import tensorflow as tf

# Text data
text = "hello world hello world"

# Character mapping
chars = sorted(set(text))

char_to_idx = {c:i for i,c in enumerate(chars)}

idx_to_char = {i:c for i,c in enumerate(chars)}

# Create sequences
seq_len = 5

X = []
y = []

for i in range(len(text)-seq_len):

    X.append([char_to_idx[c] for c in text[i:i+seq_len]])

    y.append(char_to_idx[text[i+seq_len]])

X = np.array(X).reshape(-1, seq_len, 1) / len(chars)

y = np.array(y)

# Build LSTM
model = tf.keras.Sequential([

    tf.keras.layers.LSTM(
        32,
        input_shape=(seq_len,1)
    ),

    tf.keras.layers.Dense(
        len(chars),
        activation='softmax'
    )
])

# Compile and train
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy'
)

model.fit(X, y, epochs=30, verbose=0)

# Generate text
seed = "hello"

result = seed

for _ in range(20):

    x = np.array(
        [[char_to_idx[c] for c in result[-seq_len:]]]
    ).reshape(1, seq_len, 1) / len(chars)

    pred = model.predict(x, verbose=0)

    next_char = idx_to_char[np.argmax(pred)]

    result += next_char

print(result)
```
