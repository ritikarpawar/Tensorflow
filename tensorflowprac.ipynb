```python

import tensorflow as tf

marks = tf.constant([
    [80, 75, 90],
    [70, 85, 88]
], dtype=tf.float32)

print("Student Marks:\n", marks.numpy())

total_marks_student1 = tf.reduce_sum(marks[0])
total_marks_student2 = tf.reduce_sum(marks[1])

print("\nTotal Marks of Student 1:", total_marks_student1.numpy())
print("Total Marks of Student 2:", total_marks_student2.numpy())

average_student1 = tf.reduce_mean(marks[0])
average_student2 = tf.reduce_mean(marks[1])

print("\nAverage Marks of Student 1:", average_student1.numpy())
print("Average Marks of Student 2:", average_student2.numpy())

highest_mark = tf.reduce_max(marks)
lowest_mark = tf.reduce_min(marks)

print("\nHighest Mark:", highest_mark.numpy())
print("Lowest Mark:", lowest_mark.numpy())

difference = marks[0] - marks[1]

print("\nSubject-wise Difference:")
print(difference.numpy())

weights = tf.constant([0.3, 0.3, 0.4], dtype=tf.float32)

weighted_scores = marks * weights

print("\nWeighted Scores:")
print(weighted_scores.numpy())

normalized_marks = marks / tf.reduce_max(marks)

print("\nNormalized Marks:")
print(normalized_marks.numpy())

squared_marks = tf.square(marks)

print("\nSquared Marks:")
print(squared_marks.numpy())

weights_matrix = tf.reshape(weights, (3, 1))

final_grades = tf.matmul(marks, weights_matrix)

print("\nFinal Grades:")
print(final_grades.numpy())

bonus_marks = tf.constant([[5], [5]], dtype=tf.float32)

final_result = final_grades + bonus_marks

print("\nFinal Result after Bonus:")
print(final_result.numpy())
```
